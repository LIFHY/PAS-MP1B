```python
import numpy as np

# Mengambil fitur 'age', 'blood_pressure' sebagai x_train
# Changed 'df_bb' to 'df'
x_train = np.array(df[['age', 'blood_pressure']])
# Mengambil fitur 'hypertension' sebagai y_train
# Changed 'df_bb' to 'df'
y_train = np.array(df['hypertension'])

print(f'x_train:\n{x_train}\n')
print(f'y_train: {y_train}')
#menyiapkan data input (x_train) dan target (y_train) dari DataFrame df, untuk pelatihan model pembelajaran mesin. Fitur yang digunakan adalah age dan blood_pressure, sedangkan targetnya adalah hypertension.
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train = le.fit_transform(y_train)
print(f'y_train:\n{y_train}')
#Kode ini mengubah nilai label target (y_train) dari format kategori menjadi angka menggunakan LabelEncoder. Hal ini penting untuk digunakan dalam algoritma pembelajaran mesin yang hanya menerima data numerik sebagai input.






```

```python
# Drop rows with NaN values in 'age' or 'blood_pressure'
df = df.dropna(subset=['age', 'blood_pressure'])

# Recreate x_train and y_train after removing NaN values
x_train = np.array(df[['age', 'blood_pressure']])
y_train = np.array(df['hypertension'])

# Convert y_train to numerical using LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)

# Now fit the model
from sklearn.neighbors import KNeighborsClassifier
K = 5
model = KNeighborsClassifier(n_neighbors=K)
model.fit(x_train, y_train) # This should now work without error

#Kode ini membersihkan data dari nilai kosong, menyiapkan fitur (x_train) dan target (y_train), mengonversi label kategori menjadi numerik, dan membuat model KNN dengan parameter 5 tetangga.
```

```python
import numpy as np # Import numpy
from sklearn.preprocessing import LabelEncoder # Import LabelEncoder

# Instantiate LabelEncoder
le = LabelEncoder() # Create an instance of LabelEncoder
# Mengonversi label kategori menjadi angka (misal, 'yes' menjadi 1 dan 'no' menjadi 0).


X_test = np.array([[52, 80], [49, 75], [23, 60], [44, 70]])
# Array data numerik untuk pengujian (misal, usia dan tekanan darah).


y_test = le.fit_transform(['yes', 'no', 'yes', 'no']).flatten()
# Transformasi label kategori ('yes', 'no') menjadi angka.
print(f'X_test:\n{X_test}')
print(f'y_test:\n{y_test}')
# Menampilkan array X_test dan y_test.
```

```python
y_pred = model.predict(X_test)
y_pred
# Kode ini menjalankan prediksi menggunakan model yang telah dilatih untuk menghasilkan output (y_pred) berdasarkan data uji X_test.
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
print(f'Accuracy:{acc} ')
# Kode ini menghitung dan menampilkan akurasi model berdasarkan prediksi dan nilai sebenarnya dari y_test.
```

```python
from sklearn.metrics import precision_score

# Change 'binary' to 'weighted' or any other valid option like 'micro', 'macro', or None
prec = precision_score(y_test, y_pred, average='weighted')

print(f'Precision: {prec}')
```

```python
from sklearn.metrics import recall_score
rec = recall_score(y_test, y_pred, average='weighted') # or 'micro', 'macro', or None
print(f'Recall:{rec} ')
```

```python
from sklearn.metrics import f1_score

# Specify an appropriate average method for multiclass classification
f1 = f1_score(y_test, y_pred, average='weighted') # or 'micro', 'macro', or None depending on your needs

print(f'F1-score: {f1}')
```

```python
from sklearn.metrics import classification_report

cls_report = classification_report(y_test, y_pred)

print(f'Classification Report:\n{cls_report}')
```
